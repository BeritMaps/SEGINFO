```python
import pandas as pd

df = pd.read_csv("/content/movie_statistic_dataset.csv")

# Año con más películas
print("Año con más películas:", df["production_date"].value_counts().idxmax())

# Género más común
print("Género más producido:", df["genres"].mode()[0])

# Directores fallecidos
print("Directores fallecidos:")
print(df[df["director_deathYear"].notna()][["director_name", "director_deathYear"]])

# Relación año - género
relacion = df.groupby("production_date")["genres"].agg(lambda x: x.mode()[0])
print("Género más común por año:")
print(relacion)
```

```python


runtime = df["runtime_minutes"]
rating = df["movie_averageRating"]
gross = df["Worldwide gross $"]

# Medidas de tendencia central
mean = runtime.mean()
median = runtime.median()
mode = runtime.mode()[0]

# Percentiles
p25 = runtime.quantile(0.25)
p50 = runtime.quantile(0.50)
p75 = runtime.quantile(0.75)

# Cuartiles
q1 = p25
q2 = p50
q3 = p75

# Variabilidad
data_range = runtime.max() - runtime.min()
variance = runtime.var()
std_dev = runtime.std()
iqr = q3 - q1
cv = std_dev / mean * 100

# Asociación
cov = np.cov(rating, gross, ddof=1)[0][1]
corr = rating.corr(gross)
r2 = corr**2

# Límites boxplot
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

# Mostrar resultados
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

print("P25:", p25)
print("P50:", p50)
print("P75:", p75)

print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)

print("Range:", data_range)
print("Variance:", variance)
print("Std Dev:", std_dev)
print("IQR:", iqr)
print("CV:", cv)

print("Covariance:", cov)
print("Correlation:", corr)
print("R^2:", r2)

print("Lower boxplot limit:", lower_limit)
print("Upper boxplot limit:", upper_limit)
```
